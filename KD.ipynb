```python
def kd(student_model,teacher_model,trainloader,temperature,epochs):
    teacher_model = teacher_model.to(device)
    student_model.train()  # Set the model to training mode
    optimizer = torch.optim.Adam(student_model.parameters(),lr=0.001)
    criterion = torch.nn.BCEWithLogitsLoss()
    for j in range(epochs):
        i = 0
        running_loss = 0.0
        for inputs, _ in trainloader:
            inputs = inputs.to(device)
            with torch.no_grad():
                targets = teacher_model(inputs).to(device)
                targets = F.softmax(targets / temperature, dim=1)
                targets=  targets.to(device)
            optimizer.zero_grad()  # Zero the gradients
            outputs = student_model(inputs)  # Forward pass
            loss = criterion(outputs, targets)  # Compute loss
            loss.backward()  # Backward pass
            optimizer.step()  # Update weights
            running_loss += loss.item() * inputs.size(0)
            if i %100 ==0:
                epoch_loss = running_loss / len(trainloader.dataset)
                print(f'Training Loss: {epoch_loss:.4f}')
            i = i+1
```

```python
student_model =  smallVGG().to(device)
teacher_path  = 'teacher_model.pt'
teacher_model  = myVGG().to(device)
teacher_model.load_state_dict(torch.load(teacher_path))
teacher_model = teacher_model.to(device)
kd(student_model,teacher_model,trainloader,5,EPOCHS)
```
